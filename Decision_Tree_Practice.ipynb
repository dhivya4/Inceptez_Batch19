```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/nursnaaz/DataSciencewithPractical/main/18.%20Decision%20Tree%20-%20Classification/play_tennis.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
X = data.iloc[:,1:-1]
```

```python
X = pd.get_dummies(X)
```

```python
X.head(1).values
```

```python
Y= data['play']
```

```python
data['play'].unique()
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(X,Y)
```

```python
model
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   feature_names=X.columns,  
                   class_names=['No', 'Yes'],
                   filled=True)
```

```python
model.predict(X)
```
